```python

mp = """\
.X.X......
..........
"""


mp = """\
.X.X....X.
.....X....
"""


mp = """\
.X.X....X.
.....X....
.....X....
"""




import math 
def dist(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)


def dist(a, b):
    if a == b:
        return 0
    elif a[0] == b[0] or a[1] == b[1]:
        return 1
    else:
        return 1.4142
def measure(path):
    d = 0 
    for i in range(len(path)-1):
            d = d + dist(path[i], path[i+1])
    return d






map = {}


(line, column, width, height) = (0, 0, 0, 0)
for c in mp:
    if c == '\n':
        line += 1
        width = column
        column = 0
    else:
        if c == 'X':
            pass
        else:
            location = (line, column)
            map[location] = []
        column += 1
height = line






for line in range(height):
    for column in range(width):
        location = (line, column)
        if location in map:
            for dx in [1, 0, -1]: # perhaps we could be smarter in choosing neighbors
                for dy in [1, 0, -1]: # nearest neighbor
                    neighbor = (line + dy, column + dx)
                    if neighbor != location and neighbor in map:
                        map[location] += [ neighbor ]




solution = []    
def dfs(paths, end, graph):
    global solution
    if (paths == []):
        return paths 
    else:
        newPaths = []
        for path in paths:
            if path[-1] == end:
                if solution == [] or measure(solution) > measure(path):
                    solution = path 
                    print(" Solution: ", path, " cost:", measure(solution))
                continue
            if measure(path) >= measure(solution) > 0:
                continue
            neighbors = graph[path[-1]]
            for neighbor in neighbors:
                if neighbor not in path:
                    newPaths = newPaths + [path + [neighbor]]
        return dfs(newPaths, end, graph)




# answer = sp([["G"]], "N", map)


from datetime import datetime


time_now = datetime.now().strftime("%H:%M:%S")
print("The current date and time is", time_now)


startin = [(0, 0)]
goingto = (height-1, width-1) # (4, 8)
answer = dfs([startin], goingto, map)


time_now = datetime.now().strftime("%H:%M:%S")
print("The current date and time is", time_now)


def report(map, answer=solution):
    letter = 97 
    for line in range(height):
        for column in range(width):
            location = (line, column)
            if location in answer:
                print(chr(letter), end=' ')
                letter +=1 
            elif location in map:
                print('.', end=' ')
            else:
                print('X', end=' ')
        print()
    print("Solution length: ", measure(answer))


report(map, answer = solution)
```
